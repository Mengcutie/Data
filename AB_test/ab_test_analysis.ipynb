### II - A/B 测试

由于与每个事件相关的时间戳，可以在进行每次观察时连续运行假设检验。

然而，问题的难点在于，一个页面被认为比另一页页面的效果好得多的时候就要停止检验吗？还是需要在一定时间内持续发生？需要将检验运行多长时间来决定哪个页面比另一个页面更好？

一般情况下，这些问题是A/B测试中最难的部分。 


`1.` 现在，要考虑的是，需要根据提供的所有数据做出决定。如果想假定旧的页面效果更好，除非新的页面在类型I错误率为5%的情况下才能证明效果更好，那么，零假设和备择假设是什么？（可以根据旧页面与新页面的转化率 **$p_{old}$** 与 **$p_{new}$** 来陈述假设。）

$$H_0:  p_{old}  -p_{new} \geq 0$$

$$H_1: p_{old}  -p_{new} < 0$$

`2.` 假定在零假设中，不管是新页面还是旧页面，$p_{new}$ 和 $p_{old}$ 都具有等于**转化**成功率的“真”成功率，也就是说 $p_{new}$ 与 $p_{old}$ 是相等的。此外，假设它们都等于**ab_data.csv** 中的 **转化率**，新旧页面都是如此。 <br><br>

每个页面的样本大小要与 **ab_data.csv** 中的页面大小相同。  <br><br>

执行两次页面之间**转化**差异的抽样分布，计算零假设中10000次迭代计算的估计值。  <br><br>

使用下面的单元格提供这个模拟的必要内容。<br><br>

a. 在零假设中，$p_{new}$ 的 **convert rate（转化率）** 是多少? <br><br>

```python
p_new = df2.converted.mean()
p_new
```

b. 在零假设中，$p_{old}$ 的 **convert rate（转化率）** 是多少? <br><br>

```python
p_old = df2.converted.mean()
p_old
```

c. $n_{new}$ 是多少?

e. 在零假设中， 使用$p_{new}$ 转化率模拟 $n_{new}$ 交易，并将这些 $n_{new}$ 1's and 0's 存储在 **new_page_converted**中。

```python
new_page_converted=np.random.choice([0,1], n_new, p=[1-p_new, p_new])
new_page_converted
```

f. 在零假设中， 使用$p_{old}$ 转化率模拟 $n_{old}$ 交易，并将这些 $n_{old}$ 1's and 0's 存储在 **old_page_converted**中。 

```python
old_page_converted=np.random.choice([0,1], n_old, p=[1-p_old, p_old])
old_page_converted
```

g. 在（e)与（f)中找到 $p_{new}$ - $p_{old}$ 模拟值。

```python
new_converted=new_page_converted.mean()
old_converted=old_page_converted.mean()
new_converted-old_converted
```

h. 使用**a. 到 g.** 中的计算方法来模拟10,000个 $p_{new}$ - $p_{old}$ 值，并将10000个值存储在 **p_diffs** 中。

```python
p_diffs = []
for _ in range (10000):
    new_page_converted=np.random.choice([0,1], n_new, p=[1-p_new, p_new])
    old_page_converted=np.random.choice([0,1], n_old, p=[1-p_old, p_old])
    new_ctr=new_page_converted.mean()
    old_ctr=old_page_converted.mean()
    p_diffs.append(new_ctr-old_ctr)
```

```python
diffs = np.array(p_diffs)
```

k.用文字解释一下在 **j.** 中计算出来的结果。 在科学研究中，这个值是什么？根据这个数值，新旧页面的转化率是否有区别呢？

**这个叫做P值(P value), p 值是零假设为真的情况下，支持备择假设的数据或更极端内容的概率；p值大于阈值说明零假设不应该拒绝；在这里，p值等于0.9024，0.9024>0.05, 所以不拒绝零假设，没有足够证据支持新页面效果更好。**

### III - 回归分析法之一

`1.` 在最后一部分，会看到，之前的A/B测试结果也可以通过执行回归来获取。<br><br>

a. 既然每行的值是转化或不转化，那么在这种情况下，应该执行哪种类型的回归？

可以执行逻辑回归，因为逻辑回归是实现，预测可能的两种结果。

b. 目标是使用 **statsmodels** 来拟合在 **a.** 中指定的回归模型，以查看用户收到的不同页面是否存在显著的文化差异。首先，为这个截距创建一列，并为每个用户收到的页面创建一个虚拟变量列。添加一个 **截距** 列， 一个 **ab_page** 列， 当用户接收 **treatment** 为1，**control**时为0。

```python
df2['intercept']=1
df2['ab_page'] = pd.get_dummies(df2['group'])['treatment']
```

```python
df2.head()
```

c. 使用 **statsmodels** 导入回归模型。  实例化该模型, 并使用 在**b.** 中创建的两个列来拟合该模型，用来预测一个用户是否会发生转化。

```python
import statsmodels.api as sm

log_mod = sm.Logit(df2['converted'], df2[['intercept', 'ab_page']])
results=log_mod.fit()
results.summary()
```

d. 在下方提供模型摘要。

在这里，ab_page的p值是0.190

e.与 **ab_page** 关联的p-值是多少?为什么它在 **II** 中的发现不同?<br><br>  

ab_page的p值是0.190，在这里，回归模型的零假设是 H_0:β= 0，备择假设是H_1: B ≠ 0，属于双尾检验；在Part II中，p值是0.9061，零假设是H_0: P_old - P_new≥0, 备择假设是 P_old - P_new<0, 属于单尾检验。单尾检验允许在差异相对较小时拒绝H_0，差异依赖于方向。而双尾检验需要相对较大的差异，差异不依赖于方向，所以这里ab_page的p值和part II中的p值不同。

f.现在在考虑其他可能影响用户是否发生转化的因素。讨论为什么考虑将其他因素添加到回归模型中是一个不错的主意。在回归模型中添加附加项有什么弊端吗？

因为很有可能有其他更相关的变量，对模型的分析结果更有帮助，能更好地拟合模型；加入其他变量到回归模型的缺点是可能会让变量之间有关联，而不是每个变量和一个变量有关联，导致结果不准确。

g. 现在，除了测试不同页面的转化率是否会发生变化之外，还要根据用户居住的国家或地区添加一个effect项。需要导入 **countries.csv** 数据集，并将数据集合并在适当的行上。  [Here](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.join.html) 是链接表格的文档。 

这个国家项对转化有影响吗？

```python
countries_df = pd.read_csv('countries.csv')
df_new = countries_df.set_index('user_id').join(df2.set_index('user_id'), how='inner')
```

```python
df_new.reset_index(inplace=True)
```

```python
df_new
```

```python
### Create the necessary dummy variables
df_new[['CA', 'UK', 'US']]=pd.get_dummies(df_new['country'])
df_new[['new_page', 'old_page']]=pd.get_dummies(df_new['landing_page'])
```

```python
df_new['intercept']=1
log_mod = sm.Logit(df_new['converted'], df_new[['intercept', 'US', 'UK', 'new_page']])
results=log_mod.fit()
results.summary()
```

从上面结果可以看出，以CA为基准线，US的p值0.130，UK的p值0.074，都大于阈值0.05，不拒绝零假设，没有足够证据支持新页面可以带来更多的转化率，而国家对转化率没有显著影响。

h. 虽然已经查看了国家与页面在转化率上的个体因素，但现在要查看页面与国家/地区之间的相互作用，则是其是否会对转化产生重大影响。创建必要的附加列，并拟合新的模型。

提供摘要结果，以及根据结果得出结论。

```python
df_new['new_CA']=df_new['new_page']*df_new['CA']
```

```python
df_new['new_UK']=df_new['new_page']*df_new['UK']
```

```python
df_new['new_US']=df_new['new_page']*df_new['US']
```

```python
df_new['intercept']=1
logit_mod = sm.Logit(df_new['converted'], df_new[['intercept', 'US', 'UK', 'new_page', 'new_US', 'new_UK']])
results=logit_mod.fit()
results.summary()
```

从页面和国家的相互作用结果来看，new_US的p值是0.383，new_UK的p值1.378，大于0.05，同时，US，UK 和new_page的p值都大于阈值0.05,说明不拒绝零假设，没有足够证据支持新页面效果会更好，说明页面和国家对转化率不会有显著影响。
